```python
# windspeed가 0인 데이터의 건수 출력
train.windspeed.value_counts()[0]
train.windspeed.value_counts().loc[0] # 연산이 적은 방향으로 코딩하는게 좋음, value_counts()는 0 뿐만 아니라
# 다른 값들의 개수도 세기 때문에 0만 세는 것 보다 연산이 길어짐.

len(train[train.windspeed == 0])
train[train.windspeed == 0].count()
(train.windspeed == 0).sum()
```

```python
def cym(datetime) :
    return "{0}-{1}".format(datetime.year, datetime.month)
# {0}, {1}은 각각 format함수의 인수의 위치    
    
train['ym'] = train['datetime'].apply(cym)
train[['datetime', 'ym']]

fig, (ax1, ax2) = plt.subplots(ncols = 2)
fig.set_size_inches(18, 4)
sns.barplot(data = train, x = 'year', y = 'count', ax = ax1)
sns.barplot(data = train, x = 'month', y = 'count', ax = ax2)

fig, ax3 = plt.subplots()
fig.set_size_inches(18, 4)
sns.barplot(data = train, x = 'ym', y = 'count', ax = ax3)
```

```python
# windspeed열의 0 값을 다른 값으로 대체
# - windspeed열 평균으로 대체
# - windspeed열 최소값/최대값으로 대체
# - windspeed열 weather(season, ...) 그룹화 -> 그룹별 windspeed평균 => 값으로 대체
# - windspeed값이 0인 행 데이터와 가장 유사한 행 data를 n개 추출 -> n개 데이터의 windspeed 평균 => 값으로 대체
# => train.csv로 저장 후 카페 제출
```

```python
# - windspeed열 평균으로 대체

train.windspeed.apply(lambda x : train.windspeed.mean() if x == 0 else x)
train.windspeed.replace(0, train.windspeed.mean())
```

```python
# - windspeed열 최소값/최대값으로 대체
train.windspeed.replace(0, train.windspeed.min())
train.windspeed.replace(0, train.windspeed.max())
```

```python
# - windspeed열 weather(season, ...) 그룹화 -> 그룹별 windspeed평균 => 값으로 대체

# weather 그룹화
weather_mean = train.groupby('weather')['windspeed'].mean()
```
