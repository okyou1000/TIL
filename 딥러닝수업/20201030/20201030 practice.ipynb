```python
docs = [
    '먹고 싶은 사과',
    '먹고 싶은 바나나',
    '길고 노란 바나나 바나나',
    '저는 과일이 좋아요'
]
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
N = len(docs)

def tf(t, d) :
    return d.count(t)

def idf(t) :
    df = 0
    for doc in docs :
        df += t in doc
    return log(N / (df + 1))

def tfidf(t, d) :
    return tf(t, d) * idf(t)
```

```python
result = []
for i in range(N) :
    result.append([])
    d = docs[i]
    for j in range(len(vocab)) :
        t = vocab[j]
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
result = []
for j in range(len(vocab)) :
    t = vocab[j]
    result.append(idf(t))
    
idf_ = pd.DataFrame(result, index = vocab, columns = ['idf'])
idf_
```

```python
result = []
for i in range(N) :
    result.append([])
    d = docs[i]
    for j in range(len(vocab)) :
        t = vocab[j]
        
        result[-1].append(tfidf(t, d))
        
tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

```python
import numpy as np
doc1 = np.array((2, 3, 0, 1))
doc2 = np.array((1, 2, 3, 1))
doc3 = np.array((2, 1, 2, 2))
```

```python
docInput = np.array([1, 1, 0, 1])
```

```python
# 퀴즈1.
# docInput 문서와 가장 가까운 문서를 출력(유클리디안 거리 공식)

def dist(x, y) :
    return sqrt(pow(x - y, 2).sum())
#     return (np.sqrt(np.sum((x - y)**2)))

print(dist(doc1, docInput))
print(dist(doc2, docInput))
print(dist(doc3, docInput))
```

```python
# 자카드 유사도 알고리즘
# 교집합 : 두 집합에서 공통으로 가지고 있는 원소 집합
# 합집합 : 두 집합에서 가지고 있는 원소 집합
# 합집합에서 교집합의 비율을 구하자. => 두 문서의 유사도

# J(doc1, doc2) = (doc1 교집합 doc2) / (doc1 합집합 doc2)
# 0 <= J(doc1, doc2) <= 1

# (doc1 합집합 doc2) = doc1 + doc2 - doc1교집합doc2


doc1 = "apple banana everyone like likey watch card holder"
doc2 = "apple banana coupon passport love you"
```

```python
tok1 = doc1.split()
tok2 = doc2.split()
```

```python
print(tok1)
print(tok2)
```

```python
# 합집합
union = set(tok1).union(set(tok2))
union
```

```python
# 교집합
intersection = set(tok1).intersection(set(tok2))
intersection
```

```python
# 자카드 유사도
print("자카드유사도 =", len(intersection) / len(union))
```

```python
# 퀴즈2.
doc1 = np.array((2, 3, 0, 1))
doc2 = np.array((1, 2, 3, 1))
doc3 = np.array((2, 1, 2, 2))
docInput = np.array([1, 1, 0, 1])

# cos_sim(doc1, docInput)
# cos_sim(doc2, docInput)
# cos_sim(doc3, docInput)

def cos_sim(x, y) :
    return sum(x * y) / (sqrt(sum(x**2)) * sqrt(sum(y**2)))

print(cos_sim(doc1, docInput))
print(cos_sim(doc2, docInput))
print(cos_sim(doc3, docInput))
```

```python
from numpy import dot
from numpy.linalg import norm # 벡터 크기 계산 함수

def cos_sim(a, b) :
    return dot(a, b) / (norm(a) * norm(b))

print(cos_sim(doc1, docInput))
print(cos_sim(doc2, docInput))
print(cos_sim(doc3, docInput))
```

```python
0 <= 피어슨 상관계수(x, y) = Covariance(x, y) / (std(x) * std(y)) <= 1

Covariance(x, y) = (Xi - Xu) * (Yi - Yu)

variance(x) = sum((Xi - Xu)**2) / n (모분산, 표준분산일 경우 n - 1)
std(x) = sqrt(variance(x))
```
